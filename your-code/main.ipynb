## Words

```python
words = ['play', 'filling', 'bar', 'theatre', 'easygoing', 'date', 'lead', 'that', 'story',  'island']
```

**Print every word in upper case**

```python
for word in words:
    print(word.upper())
```

**Create a new list containing only words with 5 or more letters**

```python
lista = ['playing', 'terrible', 'swimming' ]
```

**Print the first word starting with "t"**

```python
b = 'T'

result = [i for i in lista if i[0].casefold() == b.casefold()]
print(result)
```

## Numbers

**Create a list containing the square of every number from 1 to 10**

```python
squares_list = []
for i in range(1, 11):
    squares_list.append(i * i)
print(squares_list)
```

**Print a list containing the square of every odd number from 1 to 10**

```python
squares_of_odds = [x**2 for x in range(1, 11) if x % 2 != 0]

print(squares_of_odds)
```

**Create a list with the squares of all multiples of 8 below 1000**

```python
squares_of_multiplesof8 = [x**2 for x in range(1, 1000) if x % 8 == 0]

print(squares_of_multiplesof8)
```

## People

```python
people = [
    {
        "name": "Juan",
        "age": 34,
        "n_kids": 2
    },
    {
        "name": "Pepe",
        "age": 27,
        "n_kids": 0
    },
    {
        "name": "Sonia",
        "age": 41,
        "n_kids": 1
    },
    {
        "name": "Lucía",
        "age": 22,
        "n_kids": 2
    },
    {
        "name": "Leo",
        "age": 55,
        "n_kids": 5
    }
]
```

**How many people are there?**

```python
names = [d["name"] for d in people]
print(len(names))
```

**How many people have kids**?

```python
kids = [d["n_kids"] for d in people]
p_with_kids = list(filter(None, kids))
print(len(p_with_kids))
```

**How many kids do they have in total?**

```python
kids = [d["n_kids"] for d in people]
print(sum(kids))
```

**In a year's time, names ending with "a" will have an extra kid. Create a list of dictionaries with people's info in a year's time**

```python
b = "a"
for d in people:
    if d["name"].endswith(b):
       d["n_kids"] += 1

print(people)
    

```
